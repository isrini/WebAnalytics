```python
years=list(range(2015,2016))
##https://en.wikipedia.org/wiki/1976_in_film
#https://en.wikipedia.org/wiki/List_of_American_films_of_1990
##urls = ["https://en.wikipedia.org/wiki/"+str(i)+"_in_film" for i in years]    
urls = ["https://en.wikipedia.org/wiki/List_of_American_films_of_"+str(i) for i in years]    
print(urls)
```

```python
import pandas as pd
import numpy as np
from IPython.display import display # Allows the use of display() for DataFrames
import time
import pickle #To save the objects that were created using webscraping
import pprint
from lxml import html
import requests
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
movies_dict=dict()
for year,url in zip(years,urls):
    print(url)
    page = requests.get(url)
    tree = html.fromstring(page.content)
    #movies = tree.xpath('//h3+//*[contains(concat( " ", @class, " " ), concat( " ", "wikitable", " " ))]//td')
    
    #movies = tree.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "wikitable", " " ))]//td')
    movies=tree.xpath('//th+//td//*[contains(concat( " ", @class, " " ), concat( " ", "mw-redirect", " " ))]')
    movies_dict[year] = movies
```

```python
from lxml import etree
for i in movies:
     print(etree.tostring(i, pretty_print=True))
```
